```python
df['month_yr']=df['order_date'].apply( lambda x: x.strftime('%Y-%m'))
```

```python
df.head()
```

```python
df.groupby('month_yr').sum()
```

```python
trend=df.groupby('month_yr').sum()['sales'].reset_index()
```

```python
plt.figure(figsize=(15,8))
plt.plot(trend['month_yr'],trend['sales'],color='#00FFFF')
plt.xticks(rotation='vertical')
plt.show()
```
